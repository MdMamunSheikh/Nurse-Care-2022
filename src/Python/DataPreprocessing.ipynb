## Data pre-processing

### Data

```python
DATA_DIR = "Accelerometer Data/"
id = 15
df = pd.read_csv(ROOT+DATA_DIR+f"acc_user{id}.csv")
df.shape
```

```python
df.datetime = df.datetime.apply(lambda x: parser.parse(x).astimezone(pytz.UTC).replace(tzinfo = None))
df.head()
```

```python
df.to_pickle(ROOT+"Formated/Pickle/"+f"data_{id}.pkl", compression = "gzip")
```

```python
df.datetime = df.datetime.apply(lambda x: x.strftime("%H:%M:%S %d-%m-%Y"))
```

```python
df.to_csv(ROOT+"Formated/CSV/"+f"data_{id}.csv", index=False)
```

### Label

```python
DATA_DIR = "Care Record Data/"
id = 25
df = pd.read_csv(ROOT+DATA_DIR+f"train{id}.csv")
df.shape
```

```python
df = df[["user_id", "activity_type_id", "target_id", "start", "finish"]]
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.start = df.start.apply(lambda x: parser.parse(x).astimezone(pytz.UTC).replace(tzinfo = None))
df.finish = df.finish.apply(lambda x: parser.parse(x).astimezone(pytz.UTC).replace(tzinfo = None))
df.head()
```

```python
df.to_pickle(ROOT+"Formated/Pickle/"+f"label_{id}.pkl", compression = "gzip")
```

```python
df.start = df.start.apply(lambda x: x.strftime("%H:%M:%S %d-%m-%Y"))
df.finish = df.finish.apply(lambda x: x.strftime("%H:%M:%S %d-%m-%Y"))
```

```python
df.to_csv(ROOT+"Formated/CSV/"+f"label_{id}.csv", index=False)
```
